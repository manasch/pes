# Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Introductory Questions

```python
data = pd.read_csv("21.csv")
data
```

```python
# Q1

data["fatality_rate"] = (data["deaths"] / data["cases"]) * 100
data
```

```python
# Q2
# Plotting the bar graph

# country with maximum number of cases
max_cases = data["cases"].max()

# locating the row with the cases
max_ctry_cases = data.loc[data['cases'] == max_cases]
max_ctry_cases

max_ctry_cases.plot(x = "cases", y = "fatality_rate",  kind = "bar")


```

```python
# Q3

fig = plt.figure(figsize=(15, 5))

region_list = {}

# storing all the regions present
for region in data["region"]:
    if region not in region_list:
        region_list[region] = 0

# calculating total cases24h for each region
for reg in region_list.keys():
    region_list[reg] = data.loc[data["region"] == reg]["cases24h"].sum()

plt.bar(list(region_list.keys()), list(region_list.values()), width=0.5, color="red")
plt.xlabel("Regions")
plt.ylabel("Cases_24h")
plt.title("Cases in These Regions Over the Past 24 Hours")
plt.show()
```

```python
# Q4

data["population"] = (data["cases"] / data["cases100k"]) * 100000
data
```

# Task Questions

```python
# Q1

data.isnull().sum()
```

```python
# Q1
# Filling the null values

data["cases"].fillna(data["cases"].mean(), inplace=True)
data["cases"] = data["cases"].astype(int)

data["cases100k"].fillna(data["cases100k"].mean(), inplace=True)

data["cases7days"].fillna(data["cases7days"].mean(), inplace=True)
data["cases7days"] = data["cases7days"].astype(int)

data["cases7days100k"].fillna(data["cases7days100k"].mean(), inplace=True)

data["deaths"].fillna(data["deaths"].mean(), inplace=True)
data["deaths"] = data["deaths"].astype(int)

data["deaths100k"].fillna(data["deaths100k"].mean(), inplace=True)

data["deaths7days"].fillna(data["deaths7days"].mean(), inplace=True)
data["deaths7days"] = data["cases7days"].astype(int)

data["deaths7days100k"].fillna(data["deaths7days100k"].mean(), inplace=True)

data['fatality_rate'].fillna(data['deaths'] / data['cases'] * 100, inplace = True)
data['fatality_rate'].fillna(0, inplace = True)

data['population'].fillna(data['cases'] / data['cases100k'] * 100000, inplace = True)
data['population'].fillna(data['population'].mean(), inplace = True)

data
```

```python
data.isnull().sum()

# all the null values have been filled now
```

```python
# Q2

# creating a new outbreak column on a copy of the dataframe
data2 = data
data2["outbreak"] = data2["cases24h"] / data2["cases7days"]

data2 = data2.sort_values(by="outbreak", ascending=False)
data2

```

```python
# showing the critical countries only

# top 10 countries sorted by their outbreak value
data2.head(10)
```

As it is shown, only Dominica and Gabon are in a state of outbreak as their outbreak value is greater than 0.5

```python
# Q3

# Comparison study between Europe and Americas

# first plot for number of cases

reg_list = {}
for regions in data["region"]:
    if regions not in reg_list and (regions == 'Europe' or regions == 'Americas'):
        reg_list[regions] = 0

for regs in reg_list.keys():
    reg_list[regs] = data.loc[data["region"] == regs]["cases"].sum()

plt.bar(list(reg_list.keys()), list(reg_list.values()), width=0.25, color="red")
plt.xlabel("Regions")
plt.ylabel("Cases")
plt.title("Cases in Europe and Americas")
plt.show()
```

```python
# second plot for number of deaths

reg_list = {}
for regions in data["region"]:
    if regions not in reg_list and (regions == 'Europe' or regions == 'Americas'):
        reg_list[regions] = 0

for regs in reg_list.keys():
    reg_list[regs] = data.loc[data["region"] == regs]["deaths"].sum()

plt.bar(list(reg_list.keys()), list(reg_list.values()), width=0.25, color="g")
plt.xlabel("Regions")
plt.ylabel("Deaths")
plt.title("Deaths in Europe and Americas")
plt.show()
```

As it can be noted, Europe has been affected worse in both the number of deaths as well as the number of cases, than the Americas by COVID, hence the statement is TRUE
