```python
#Task Question 1
#finding which columns have a null value
df.isnull().sum()
```

```python
#identifying and displaying all the rows which have a null value
isnan=df.isnull()
rownan=isnan.any(axis=1)
rowswithnan=df[rownan]
rowswithnan
```

```python
df.fillna({'cases':df['cases'].mean(),'cases100k':df['cases100k'].mean(),'cases7days':df['cases7days'].mean(),
           'cases7days100k':df['cases7days100k'].mean(),'deaths':df['deaths'].mean(),
           'deaths100k':df['deaths100k'].mean(),'deaths7days':df['deaths7days'].mean(),'deaths7days100k':df['deaths7days100k'].mean(),
           'Fatality Rate':df['Fatality Rate'].mean(),'Population':df['Population'].mean()}, inplace=True)
```
